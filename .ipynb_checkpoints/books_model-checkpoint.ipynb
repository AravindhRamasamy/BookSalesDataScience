# Data Preprocessing #

```python
# create a function to transform or convert the year into date before extrating it
date_formats = ['%Y-%m-%d %H:%M:%S.%f', '%Y']
# Function to parse dates with multiple formats
def try_parsing_date(text):
    for fmt in date_formats:
        try:
            return pd.to_datetime(text, format=fmt)
        except ValueError:
            continue
    return pd.NaT  # Return NaT (Not a Time) if none of the formats match

# Apply the function to the 'dates' column
df['Date'] = df['Year'].apply(try_parsing_date)
```

```python
df
```

```python
# drop the year column
df.drop('Year', axis=1, inplace=True)
df
```

```python
#extract the year from the date
extract = pd.DatetimeIndex(df['Date'])
df['Year'] = extract.year
df['Year'] = df['Year'].astype(str).str.extract(r'(\d+)')
df
```

```python
# Drop the date column
df.drop('Date', axis=1, inplace=True)
```

```python
# replace the null column by some data
df['category_name'].fillna('Non_define', inplace = True)
df['Genre'].fillna('Non_define', inplace = True)
df['Author'].fillna('Non_define', inplace = True)
df['Year'].fillna(df['Year'].median, inplace = True)
df
```

```python
X = df[['User Rating','Reviews']]
y = df['Price']
```

```python
X_train,X_test,y_train,y_test = sklearn.model_selection.train_test_split(X,y, test_size = 0.1,random_state = 0)
print(f'Size of the dataset : {df.shape}')
print(f'size of X_train : {X_train.shape}')
print(f'size of X_test : {X_test.shape}')
```

# Feature Engineering #

```python
num_attrib = X_train.select_dtypes(include = ['int64', 'float64']).columns.tolist()
categorical_attributes = X_train.select_dtypes(include='object').columns.tolist()
column_transform = sklearn.compose .ColumnTransformer([
    ('standard_scalling',sklearn.preprocessing.StandardScaler(), num_attrib),
    ('one_hot_encoding',sklearn.preprocessing.OneHotEncoder(handle_unknown='ignore'), categorical_attributes),
], remainder = 'passthrough')
column_transform.fit(X_train)
X_train = column_transform.transform(X_train)
X_test = column_transform.transform(X_test)
```

```python
print(f'Dataset size : {df.shape}')
print(f'X_train size : {X_train.shape}')
print(f'X_test size : {X_test.shape}')
```

# Model selection and hyperparameter tuning #

```python
param_grid = {
   'fit_intercept': [True],
   'copy_X': [True],
   'positive': [True]
}
```

```python
Linear = sklearn.linear_model.LinearRegression()
```

```python
modelLinear = sklearn.model_selection.GridSearchCV(Linear,param_grid, scoring = 'accuracy', cv=5,n_jobs=1)
modelLinear.fit(X_train,y_train)
best_params = modelLinear.best_params_
best_estimato = modelLinear.best_estimator_
print("Best Parameters:", best_params)
print(f'Decision Tree best estimator : {best_estimato}')
```
